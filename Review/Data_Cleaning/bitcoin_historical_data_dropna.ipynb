# Use dropna to cleaned missing values.


```python
# Import libraries
import pandas as pd
```

```python
# specify the path to data file
data = 'bitstamp.csv'

# Load the data into a Data Frame
df = pd.read_csv(data)
```

```python
# Show the first 5 rows
df.head()
```

```python
# Check information and missing values
df.info()
```

```python
# Drop the rows where at least one element is missing value
df.dropna()
```

```python
# Drop the columns where at least one element is missing
df.dropna(axis='columns')
```

```python
# Drop the rows where all elements are missing
df.dropna(how='all')
```

```python
# Keep only the rows with at least 2 non-NA values
df.dropna(thresh=2)
```
