# Remove missing values with DataFrame.dropna



```python
#Import bibliotecas
import pandas as pd
import numpy as np
```

```python
#Criando data example
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                    "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                             pd.NaT]})
```

```python
#Drop the rows where at least one element is missing
#Elimine as linhas em que falta pelo menos um elemnto.

df.dropna()
```

```python
#Drop the columns where at least one element is missing.
#Eliminar as colunas em que falta pelo menos um elemento.

df.dropna(axis='columns')
```

```python
#Drop the rows where all elements are missing.
#Elimine linhas em que faltam todos os elementos

df.dropna(how='all')
```

```python
#Keep only the rows with at least 2 non-NA values.
#Manter apenas as linhas com pelo menos 2 valores não-NA.

df.dropna(thresh=2)
```

```python
#Define in which columns to look for missing values.
#Difina a coluna para procurar por valores perdidos.

df.dropna(subset=['name','toy'])
```
