```python
import pandas as pd
import numpy as np
```

```python
# Load Path
df = pd.read_csv('/kaggle/input/zomato/zomato.csv')
```

```python
# Show the  first rows
df.head()
```

# Checking information and missing values


```python
df.info()
```

```python
# Looking for null values
num_values = df.isnull().sum()
print(num_values)
```

Columns with null values:
* rate
* phone
* location
* rest_type
* dish_liked
* cuisines
* approx_cost(for tow people)


```python
# Clean rate column
print(df['rate'].head())

# Information about the column
df['rate'].info()

# Null values
rate_null_values = df['rate'].isnull().sum()
print(rate_null_values)

# I could turn it into numbers and make an average, but i don't think it's relevant.
# Delete null values
df['rate'].dropna(inplace= True)
```

```python
# phone column
df['phone'].head()

# Check null values
df['phone'].isnull().sum()

# For phone numbers I can just delete the missing values or fill in zero

# Fill null values with zero
df['phone'].fillna(0, inplace=True)

# Remove non-numeric characters and whitespace
df['phone'] = df['phone'].replace(to_replace=r'[^+\d\\]', value='', regex=True)
```

```python
# Location column 

# Show the first five rows
df['location'].head()

# Check for null values
df['location'].isnull().sum()

# Delete null values
df['location'].dropna(inplace = True)
```

```python
# rest_type column

# Show the first five rows
df['rest_type'].head()

# Check for null values
df['rest_type'].isnull().sum()

# Drop null values
df['rest_type'].dropna(inplace=True)
```

```python
# dish_liked

# Show the first five rows
df['dish_liked'].head()

# Remove leading and trailing whitespaces in the column
df['dish_liked'] = df['dish_liked'].str.strip()

# Standardize capitalization 
df['dish_liked'] = df['dish_liked'].str.upper()

# Drop null values in the dishe_liked
df.dropna(subset=['dish_liked'], inplace=True)
print(df['dish_liked'])
```

```python
# cuisines

# Show the first 5 columns
df['cuisines'].head()

# Drop null values
df['cuisines'].isnull().sum
df.dropna(subset=['cuisines'], inplace= True)
```

```python
#approx_cost(for tow people)

# Show the five first rows
df['approx_cost(for two people)'].head()

# Check for null values
df['approx_cost(for two people)'].isnull().sum()

# Replace null values with average
df['approx_cost(for two people)'].info()

# Convert to float
df['approx_cost(for two people)'] = pd.to_numeric(df['approx_cost(for two people)'], errors='coerce')

# Display resulting DataFrame
print(df['approx_cost(for two people)'])

# Replace null values with average
df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].mean(), inplace=True)
```

# General data cleaning

```python
# Cleaning data
df.isnull().sum()
df.dropna(inplace=True)
```

```python
# Standardize capitalization 
df = df.applymap(lambda x: x.upper()if isinstance(x, str) else x)

#  Remove WitheSpace
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Show the DataFrame
print(df)
```

```python
# Remove duplicate values
df.drop_duplicates(inplace=True)

# Show the dataFrame
df.head()
```
